```python
# =========================
# IMPORT LIBRARIES
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from google.colab import files # Only needed if we resort to manual upload
import requests # For downloading files from URL

# =========================
# LOAD DATA
# =========================
file_name = "StudentsPerformance.csv"
url = "https://raw.githubusercontent.com/plotly/datasets/master/Dash-Course/StudentsPerformance.csv"
df = None # Initialize df to None

# Attempt to load the file
if os.path.exists(file_name):
    print(f"Loading '{file_name}' from local storage.")
    df = pd.read_csv(file_name)
else:
    print(f"File '{file_name}' not found locally. Attempting to download from URL.")
    try:
        response = requests.get(url)
        response.raise_for_status() # Raise an exception for HTTP errors
        with open(file_name, 'wb') as f:
            f.write(response.content)
        print(f"Successfully downloaded '{file_name}' from URL.")
        df = pd.read_csv(file_name)
    except requests.exceptions.RequestException as e:
        print(f"Failed to download '{file_name}' from URL: {e}")
        print("Please upload the file manually.")
        uploaded = files.upload()
        if file_name in uploaded:
            print(f"'{file_name}' uploaded successfully.")
            df = pd.read_csv(file_name)
        else:
            print(f"'{file_name}' was not uploaded or name mismatch. Please ensure the correct file is selected.")
            df = None # Ensure df remains None if upload fails or is cancelled
    except Exception as e:
        print(f"An unexpected error occurred during download or initial load: {e}")
        print("Please upload the file manually.")
        uploaded = files.upload()
        if file_name in uploaded:
            print(f"'{file_name}' uploaded successfully.")
            df = pd.read_csv(file_name)
        else:
            print(f"'{file_name}' was not uploaded or name mismatch. Please ensure the correct file is selected.")
            df = None


if df is not None:
    print("First 5 Rows:")
    print(df.head())

    print("\nShape of Dataset:")
    print(df.shape)

    print("\nData Types:")
    print(df.dtypes)

    # =========================
    # DATA CLEANING
    # =========================
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Remove duplicates
    df.drop_duplicates(inplace=True)

    print("\nAfter Removing Duplicates Shape:")
    print(df.shape)

    # =========================
    # STATISTICS (MATH & READING)
    # =========================

    print("\n--- MATH SCORE STATS ---")
    print("Mean:", df['math score'].mean())
    print("Median:", df['math score'].median())
    print("Mode:", df['math score'].mode()[0])
    print("Standard Deviation:", df['math score'].std())
    print("Variance:", df['math score'].var())

    math_range = df['math score'].max() - df['math score'].min()
    math_midrange = (df['math score'].max() + df['math score'].min()) / 2

    print("Range:", math_range)
    print("Mid-range:", math_midrange)

    print("\n--- READING SCORE STATS ---")
    print("Mean:", df['reading score'].mean())
    print("Median:", df['reading score'].median())
    print("Mode:", df['reading score'].mode()[0])
    print("Standard Deviation:", df['reading score'].std())
    print("Variance:", df['reading score'].var())

    # =========================
    # VISUALIZATIONS
    # =========================

    # Histogram
    plt.figure()
    plt.hist(df['math score'], bins=20)
    plt.title("Distribution of Math Scores")
    plt.xlabel("Math Score")
    plt.ylabel("Frequency")
    plt.show()

    # Countplot
    plt.figure()
    sns.countplot(x='gender', data=df)
    plt.title("Gender Distribution")
    plt.xlabel("Gender")
    plt.ylabel("Count")
    plt.show()

    # Boxplot
    plt.figure()
    sns.boxplot(x=df['math score'])
    plt.title("Math Score Spread and Outliers")
    plt.show()

    # Heatmap
    plt.figure()
    sns.heatmap(df[['math score','reading score','writing score']].corr(), annot=True)
    plt.title("Correlation Heatmap")
    plt.show()

    # =========================
    # ENCODING
    # =========================
    from sklearn.preprocessing import LabelEncoder

    le = LabelEncoder()

    # Check if columns exist before encoding, as some might be missing in a malformed CSV
    if 'gender' in df.columns:
        df['gender'] = le.fit_transform(df['gender'])
    if 'lunch' in df.columns:
        df['lunch'] = le.fit_transform(df['lunch'])
    if 'test preparation course' in df.columns:
        df['test preparation course'] = le.fit_transform(df['test preparation course'])

    # =========================
    # TRAIN TEST SPLIT
    # =========================
    from sklearn.model_selection import train_test_split

    # Ensure required columns for X and y exist
    required_cols_X = ['reading score','writing score']
    required_col_y = 'math score'

    if all(col in df.columns for col in required_cols_X) and required_col_y in df.columns:
        X = df[required_cols_X]
        y = df[required_col_y]

        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

        # =========================
        # MODEL
        # =========================
        from sklearn.linear_model import LinearRegression

        model = LinearRegression()
        model.fit(X_train, y_train)

        accuracy = model.score(X_test, y_test)

        print("\nModel Accuracy:", accuracy)
    else:
        print("\nError: Missing required columns for training the model. Please check the dataset.")

else:
    print("DataFrame 'df' was not loaded. Cannot proceed with analysis.")
```
